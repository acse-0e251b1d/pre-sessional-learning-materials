1. 自变量的数量： 仅涉及一个自变量的导数为ODE
                 涉及多个自变量的导数为PDE
2. 线性： 只能出现函数本身，函数任何阶次的导函数
         函数本身及所有的导函数除了加减外不能有任何运算
         函数本身和各阶导数本身不能有任何加减外的运算
         不允许对函数本身、各阶导数做任何复合运算

3. 阶： 最高导数的阶

对于一阶非齐次偏微分方程

$ 
u' + p\left( x \right)u = r\left( x \right)\
$
1. 先求解齐次方程的通解（分离变量）
2. 再将通解中的C写成c(x)的形式，将c(x)求解出来，从初始条件开始积分
3. 将c(x)带入非齐次问题中
4. 寻找非齐次解：齐通解+非齐特解

\begin{array}{l}
u\left( x \right) = {e^{ - h}}\left( {\int {{e^h}rdx + c} } \right)\\
h = \int {p\left( x \right)} dx
\end{array}




\begin{array}{l}
{u_h}\left( x \right) = {e^{ - \int_{{x_0}}^x {p(s) \cdot ds} }}\\
{u_p}\left( x \right) = c\left( x \right){e^{ - \int_{{x_0}}^x {p(s) \cdot ds} }}\\
c\left( x \right) = \int_{{x_0}}^x {r(s){e^{\int_{{x_0}}^s {p(\sigma ) \cdot d\sigma } }}ds} 
\end{array}



\begin{array}{l}
u(x) = {u_p} + {u_h}\\
 = {e^{ - \int_{{x_0}}^x {p(s) \cdot ds} }}\left( {\int_{{x_0}}^x {r(s){e^{\int_{{x_0}}^s {p(\sigma ) \cdot d\sigma } }}ds}  + C} \right)
\end{array}
其中带入初始条件C=u(x0)
