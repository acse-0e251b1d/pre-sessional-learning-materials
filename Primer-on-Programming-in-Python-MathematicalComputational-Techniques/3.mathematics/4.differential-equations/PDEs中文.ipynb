## 链式法则
## 辅助方程
初始条件； 边界条件（迪利克雷边界和诺伊曼边界）
## 适定性
物理模型及数学模型应该具有以下性质：
1. 存在性
2. 唯一性
3. 稳定性：解的行为随初始条件变化
   
## 椭圆、抛物线和双曲线 PDE
椭圆：特征根同正同负

双曲：特征根一正一负

抛物线：一个特征值为0

1. 拉普拉斯方程
${\nabla ^2}u = 0$

 2.热方程（扩散方程）
${u_t} - k{\nabla ^2}u = 0,k > 0$

3. 波动方程
${u_{tt}} - {c^2}{\nabla ^2}u = 0$

## 解的特征
椭圆：解决方案平滑，平滑度由PDE中的系数决定，边界可能粗糙，随着时间（时间是自变量）变得平滑。如果两个自变量是空间坐标，远离边界会平滑

抛物线： 通过适量的变量变化，总是可以转化为扩散方程的等效形式，随着时间增加解更平滑

双曲线： 解不会随着时间更加平滑，初始条件的不连续性往往会跨越空间和时间传播。双曲方程通常很难处理，随着时间和距离增加，解仍然复杂

# 一阶线性偏微分方程
## 特征线法：
将PDE转化为ODE
使x,y是s的函数，利用链式法则替换

$a(x,y){u_x} + b(x,y){u_y} = f(x,y,u)$

$\frac{{dx}}{{ds}} = a,\frac{{dy}}{{ds}} = b,\frac{{du}}{{ds}} = f$替换上式的a，b。变成全微分

$\frac{{du}}{{ds}} = \frac{{\partial u}}{{\partial x}}\frac{{dx}}{{ds}} + \frac{{\partial u}}{{\partial y}}\frac{{dy}}{{ds}} = f\left( {x(s),y(s),u(s)} \right)$
### 常系数齐次PDEs
a,b是常数，f==0
带入得到常微分方程-->得到dx,dy的关系（特征线）平行

将u改写成单变量函数（PDE-->ODE）
u(x,y),x,y在直线上
带入初始条件：u(x,0)=cosx

```python

from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 201)
y = np.linspace(-5, 5, 201)
X, Y = np.meshgrid(x, y)

f = np.cos(X - 2/3 * Y)

fig = plt.figure(figsize=(10, 8))

ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122)

ax1.plot_surface(X, Y, f, rstride=1, cstride=1, cmap='inferno', antialiased=False)
ax2.contourf(f, cmap='inferno', extent=[-5, 5, -5, 5])
ax2.contour(f, colors='k', extent=[-5, 5, -5, 5], linewidths=1.)

ax1.view_init(40, 70)
ax2.set_aspect('equal')

for ax in (ax1, ax2):
    ax.set_xlabel('x')
    ax.set_ylabel('y')
ax1.set_zlabel('u')
    
fig.suptitle(r'$u(x,y) = \cos (x - \frac{2}{3}y)$')

plt.tight_layout()
plt.show()
```

# [二阶线性偏微分方程的典型形式](https://blog.csdn.net/SanyHo/article/details/105977566)
$a{u_{xx}} + b{u_{xy}} + c{u_{yy}} = f\left( {x,y,u,{u_x},{u_y}} \right)$

## 椭圆方程：${b^2} - 4ac > 0$

拉普拉斯方程（a=1,b=0,c=1）
$\frac{{{\partial ^2}\phi }}{{\partial {x^2}}} + \frac{{{\partial ^2}\phi }}{{\partial {y^2}}} = 0$

泊松方程(a=1,b=0,c=1)
$\frac{{{\partial ^2}\phi }}{{\partial {x^2}}} + \frac{{{\partial ^2}\phi }}{{\partial {y^2}}} = s(x,y)$

## 抛物线：${b^2} - 4ac = 0$

扩散方程(a=1,b=0,c=0)
$\alpha {\frac{{\partial \phi }}{{\partial t}}^2} = \alpha \frac{{{\partial ^2}\phi }}{{\partial {x^2}}}$

## 双曲线：${b^2} - 4ac > 0$
   
波动方程(a=1,b=0,c=-1)
$\frac{{{\partial ^2}\phi }}{{\partial {t^2}}} - {c^2}\frac{{{\partial ^2}\phi }}{{\partial {x^2}}} = 0$

x,y,x^2-y^2,x^3-3xy^2，都是u=0的解。画出图像：

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-1, 1, 101)
X, Y = np.meshgrid(x, x)

fig = plt.figure(figsize=(10, 5))

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d')

ax1.plot(x, [0.25]*len(x), label=r'u = 0.25')
ax1.plot(x, x, label=r'$u = x$')
# ax2.imshow(X**3 - 3*X*Y**2, cmap='inferno', extent=[min(x), max(x), min(x), max(x)])
ax2.plot_surface(X, Y, X**3 - 3*X*Y**2, rstride=1, cstride=1, cmap='inferno', antialiased=False)

ax2.set_title(r'$u=x^3 - 3xy^2$')
ax1.legend(loc='best')
ax1.set_aspect('equal')
plt.show()
```

```python
def f_polar(r, theta):
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    return x**3 - 3*x*y**2


r = np.linspace(0, 0.001, 101)
theta = np.linspace(0, 2*np.pi, 360)
rr, tt = np.meshgrid(r, theta)

u = f_polar(rr, tt)

fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')

ax.pcolormesh(tt, rr, u, cmap='inferno')

ax.set_aspect('equal')
ax.set_rgrids([])
plt.show()

print('mean(u) - u(0,0) = ', np.mean(u) - f_polar(0, 0))
```

```python
x = np.linspace(-7.5, 7.5, 300)
X, Y = np.meshgrid(x, x)
r = np.sqrt(X**2 + Y**2)

u = np.sin(r)/r

fig = plt.figure(figsize=(14, 7))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122)

ax1.plot_surface(X, Y, u, rstride=1, cstride=1, cmap='inferno', antialiased=False)
ax2.imshow(u, extent=[min(x), max(x), min(x), max(x)], cmap='inferno')

ax1.set_zlabel('u')
for ax in (ax1, ax2):
    ax.set_xlabel('x')
    ax.set_ylabel('y')

ax2.set_aspect('equal')
plt.show()
```

```python
u_x = X*np.cos(r) / r**2 - X*np.sin(r) / r**3  # x-component of grad u
u_y = Y*np.cos(r) / r**2 - Y*np.sin(r) / r**3  # y-component of grad u
lapl_u = (1 - r**2) * np.sin(r) / r**3 - np.cos(r) / r**2  # laplacian u

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

im = ax.imshow(lapl_u, extent=[min(x), max(x), min(x), max(x)], cmap='winter')
ax.quiver(X[::15, ::15], Y[::15, ::15], u_x[::15, ::15], u_y[::15, ::15], color='white', scale=10, units='width')
ax.contour(np.sign(lapl_u), 2, colors='white', extent=[min(x), max(x), min(x), max(x)], alpha=0.3)

plt.colorbar(im, shrink=0.75, label=r'$\Delta u$')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 5, 201)

u = np.sin(x**2)
u_xx = 2*np.cos(x**2) - 4*x**2*np.sin(x**2)
u_xx /= u_xx.max()

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)

ax.plot(x, u, label=r'$\sin x^2$')
ax.quiver(x[::2], u[::2]*0, 0, u_xx[::2], units='xy', width=0.01)

ax.legend(loc='best')
ax.axis('off')
plt.show()
```
